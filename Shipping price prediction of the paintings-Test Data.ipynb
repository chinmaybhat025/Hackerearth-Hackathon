## As the Numerical Features which we have Imputed the median makes the variable somewhat skewed so we will further apply transformations on the same.

```python
import scipy.stats as stat
import pylab
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
test_data['Height_median_sqaure']=test_data.Height_median**(1/2)
plot_data(test_data,'Height_median_sqaure')
```

```python
test_data.drop('Height_median',axis=1,inplace=True)
```

## Handling Categorical Missing Values

```python
def impute_nan(df,variable):
    most_frequent_category=df[variable].mode().index[0]
    df[variable].fillna(most_frequent_category,inplace=True)
```

```python
for feature in ['Material','Remote Location','Transport']:
    impute_nan(test_data,feature)
```
